```python
# Size of one plot and resolution
pltSizex = 10
pltSizey = 10
dpi = 300
# Fontsize for the different parameters
legFont = 55
xTitleFont = 60
yTitleFont = 60
xLabelFont = 55
yLabelFont = 55
# Define linewidth for line plots
lw = 5
# Color for each condition
cmap = colmap.get_cmap('nipy_spectral')
amp15_col = cmap(0.0)
amp30_col = cmap(0.9)
amp45_col  = cmap(0.15)
amp60_col  = cmap(0.5)
# Error bars
ci = "sd"
```

```python
## Load robot data
all_rob_data = pd.read_csv("./all_params/Transp_2DDL_rob_cut-off-rel-5.csv", encoding = "ISO-8859-1", engine='python')
all_rob_data = all_rob_data[all_rob_data['condition'] != 'NE']
all_rob_data_fw = all_rob_data[all_rob_data['direction'] == 'forward']
all_rob_data_bw = all_rob_data[all_rob_data['direction'] == 'backward']
```

```python
## Compute aggregated interaction forces for forward moves
forward_agg_trajs = figures.aggregate_interEfforts_trajectories(all_rob_data_fw)
```

```python
## Compute aggregated interaction forces for forward moves
backward_agg_trajs = figures.aggregate_interEfforts_trajectories(all_rob_data_bw)
```

```python
# List of conditions
list_cond = ['OL','CL','OLCL']
t_pc = 100
t_deb_fin_pc = 50

fzA_dict_cond_plt = {}
fzFA_dict_cond_plt = {}
normA_dict_cond_plt = {}
normFA_dict_cond_plt = {}

for cond in list_cond:
    fzA_dict_cond = forward_agg_trajs.get('fzA').get(cond)
    fzFA_dict_cond = forward_agg_trajs.get('fzFA').get(cond)
    normA_dict_cond = forward_agg_trajs.get('normA').get(cond)
    normFA_dict_cond = forward_agg_trajs.get('normFA').get(cond)
    
    fzA_dict_upd = {}
    fzFA_dict_upd = {}
    normA_dict_upd = {}
    normFA_dict_upd = {}
    
    for targ in range(1,6):
        fzA_dict_targ = fzA_dict_cond.get(targ)
        fzFA_dict_targ = fzFA_dict_cond.get(targ)
        normA_dict_targ = normA_dict_cond.get(targ)
        normFA_dict_targ = normFA_dict_cond.get(targ)
        
        # Compute average fz at the arm on all participants
        t, fzA, fzA_born_inf, fzA_born_sup = mean_profiles(np.array(fzA_dict_targ.get('mvt')),t_pc)
        t_deb, fzA_deb, fzA_deb_born_inf, fzA_deb_born_sup = mean_profiles(np.array(fzA_dict_targ.get('deb')),t_deb_fin_pc)
        t_fin, fzA_fin, fzA_fin_born_inf, fzA_fin_born_sup = mean_profiles(np.array(fzA_dict_targ.get('fin')),t_deb_fin_pc)
        
        # Compute average fz at the forearm on all participants
        _, fzFA, fzFA_born_inf, fzFA_born_sup = mean_profiles(np.array(fzFA_dict_targ.get('mvt')),t_pc)
        _, fzFA_deb, fzFA_deb_born_inf, fzFA_deb_born_sup = mean_profiles(np.array(fzFA_dict_targ.get('deb')),t_deb_fin_pc)
        _, fzFA_fin, fzFA_fin_born_inf, fzFA_fin_born_sup = mean_profiles(np.array(fzFA_dict_targ.get('fin')),t_deb_fin_pc)
        
        # Compute average norm at the arm on all participants
        _, normA, normA_born_inf, normA_born_sup = mean_profiles(np.array(normA_dict_targ.get('mvt')),t_pc)
        _, normA_deb, normA_deb_born_inf, normA_deb_born_sup = mean_profiles(np.array(normA_dict_targ.get('deb')),t_deb_fin_pc)
        _, normA_fin, normA_fin_born_inf, normA_fin_born_sup = mean_profiles(np.array(normA_dict_targ.get('fin')),t_deb_fin_pc)
        
        # Compute average norm at the forearm on all participants
        _, normFA, normFA_born_inf, normFA_born_sup = mean_profiles(np.array(normFA_dict_targ.get('mvt')),t_pc)
        _, normFA_deb, normFA_deb_born_inf, normFA_deb_born_sup = mean_profiles(np.array(normFA_dict_targ.get('deb')),t_deb_fin_pc)
        _, normFA_fin, normFA_fin_born_inf, normFA_fin_born_sup = mean_profiles(np.array(normFA_dict_targ.get('fin')),t_deb_fin_pc)
        
        # Save average fz force at the arm per condition
        fzA_dict_upd.update({targ: {'t_plt': t, 't_deb_plt': t_deb, 't_fin_plt': t_fin,
                                    'deb_plt': fzA_deb, 'mvt_plt': fzA, 'fin_plt': fzA_fin, 'sup_plt': fzA_born_sup, 'inf_plt': fzA_born_inf}})
        # Save average fz force at the forearm per condition
        fzFA_dict_upd.update({targ: {'t_plt': t, 't_deb_plt': t_deb, 't_fin_plt': t_fin,
                                     'deb_plt': fzFA_deb, 'mvt_plt': fzFA, 'fin_plt': fzFA_fin, 'sup_plt': fzFA_born_sup, 'inf_plt': fzFA_born_inf}})
        # Save average norm at the arm per condition
        normA_dict_upd.update({targ: {'t_plt': t, 't_deb_plt': t_deb, 't_fin_plt': t_fin,
                                      'deb_plt': normA_deb, 'mvt_plt': normA, 'fin_plt': normA_fin, 'sup_plt': normA_born_sup, 'inf_plt': normA_born_inf}})
        # Save average norm at the forearm per condition
        normFA_dict_upd.update({targ: {'t_plt': t, 't_deb_plt': t_deb, 't_fin_plt': t_fin,
                                       'deb_plt': normFA_deb, 'mvt_plt': normFA, 'fin_plt': normFA_fin, 'sup_plt': normFA_born_sup, 'inf_plt': normFA_born_inf}})
    
    # Save all data for plots
    fzA_dict_cond_plt.update({cond : fzA_dict_upd})
    fzFA_dict_cond_plt.update({cond : fzFA_dict_upd})
    normA_dict_cond_plt.update({cond : normA_dict_upd})
    normFA_dict_cond_plt.update({cond : normFA_dict_upd})
```

```python
def set_axis_figInterForcesTrajs(fig,axes):
    
    axes[0,0].set_xlabel('', fontsize = yTitleFont)
    axes[0,0].set_ylabel('$F_{z,A}$ ($N$)', fontsize = yTitleFont)
    axes[0,0].set_yticks([-20,-10,0,10])
    axes[0,0].set_yticklabels([-20,-10,0,10],fontsize = yLabelFont)
    axes[0,0].set_ylim([-25,11])
    axes[0,0].set_title('$\mathrm{OL}$', fontsize = 1.2 * yTitleFont)
    
    axes[0,1].set_xlabel('', fontsize = yTitleFont)
    axes[0,1].set_ylabel('', fontsize = yTitleFont, labelpad = 10)
    axes[0,1].set_title('$\mathrm{CL}$', fontsize = 1.2 * yTitleFont)
    
    axes[0,2].set_xlabel('', fontsize = yTitleFont)
    axes[0,2].set_ylabel('', fontsize = yTitleFont, labelpad = 10)
    axes[0,2].set_title('$\mathrm{OLCL}$', fontsize = 1.2 * yTitleFont)
    
    axes[1,0].set_xlabel('Time ($\% \mathrm{MD}$)', fontsize = yTitleFont)
    axes[1,0].set_xticks([0,50,100])
    axes[1,0].set_xticklabels([0,50,100],fontsize = yLabelFont)
    axes[1,0].set_xlim([-5,105])
    axes[1,0].set_ylabel('$F_{z,FA}$ ($N$)', fontsize = yTitleFont)
    axes[1,0].set_yticks([-20,-10,0,10,20])
    axes[1,0].set_yticklabels([-20,-10,0,10,20],fontsize = yLabelFont)
    axes[1,0].set_ylim([-13,21])
    
    axes[1,1].set_xlabel('Time ($\% \mathrm{MD}$)', fontsize = yTitleFont)
    axes[1,1].set_xticks([0,50,100])
    axes[1,1].set_xticklabels([0,50,100],fontsize = yLabelFont)
    axes[1,1].set_xlim([-5,105])
    
    axes[1,2].set_xlabel('Time ($\% \mathrm{MD}$)', fontsize = yTitleFont)
    axes[1,2].set_xticks([0,50,100])
    axes[1,2].set_xticklabels([0,50,100],fontsize = yLabelFont)
    axes[1,2].set_xlim([-5,105])
    
    for axs in axes:
        for ax in axs:
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.spines['left'].set_linewidth(4)
            ax.spines['bottom'].set_linewidth(4)
            ax.tick_params(width = 4, length = 12)
            ax.grid(axis = 'y', which = 'both', lw = 1, zorder = 0)
            ax.tick_params(axis='x', which='major', pad=5)
            ax.grid(axis='y', zorder = 0)
            ax.legend().remove()
```

```python
lever = 0.3
## Define conditions list and dicts
cond_list = ['OL','CL','OLCL']
cond_dict = {'OL': 0, 'CL': 1, 'OLCL': 2}
targ_colors = {1: 'lightgrey', 2: 'darkgrey', 3: 'grey', 4: 'dimgrey', 5: 'k'}

## Create subplots
fig, axes = plt.subplots(2,3,figsize = (3 * pltSizex, 2 * pltSizey), sharex = True, sharey = 'row', dpi = dpi)

## Perform plots
for cond in cond_list:
    fzA_loc_plt = fzA_dict_cond_plt.get(cond)
    fzFA_loc_plt = fzFA_dict_cond_plt.get(cond)
    
    for targ in range(1,6):
        fzA = fzA_loc_plt.get(targ).get('mvt_plt')
        fzFA = fzFA_loc_plt.get(targ).get('mvt_plt')
        
        fzA_inf = fzA_loc_plt.get(targ).get('inf_plt')
        fzFA_inf = fzFA_loc_plt.get(targ).get('inf_plt')
        
        fzA_sup = fzA_loc_plt.get(targ).get('sup_plt')
        fzFA_sup = fzFA_loc_plt.get(targ).get('sup_plt')
        
        t = fzA_loc_plt.get(targ).get('t_plt')
        
        axes[0,cond_dict.get(cond)].plot(t, fzA, c = targ_colors.get(targ), linewidth = lw)
        axes[0,cond_dict.get(cond)].fill_between(t, fzA_inf, fzA_sup, color = targ_colors.get(targ), linewidth = lw, alpha = 0.2)
        
        axes[1,cond_dict.get(cond)].plot(t, fzFA, c = targ_colors.get(targ), linewidth = lw)
        axes[1,cond_dict.get(cond)].fill_between(t, fzFA_inf, fzFA_sup, color = targ_colors.get(targ), linewidth = lw, alpha = 0.2)


## Finalize figure
set_axis_figInterForcesTrajs(fig,axes)
fig.tight_layout(pad = 1.0)

## Save figure
plt.savefig('./Figures_Unprocessed/Av_InterForcesTrajs_fw.svg', format='svg')
```
