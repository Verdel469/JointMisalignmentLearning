```python
## Load population emgs
all_emg_data = pd.read_csv("./all_params/Transp_2DDL_emg_cut-off-rel-5.csv", encoding = "ISO-8859-1", engine='python')
all_qtm_data = pd.read_csv("./all_params/Transp_2DDL_index_cut-off-rel-5.csv", encoding = "ISO-8859-1", engine='python')
```

```python
all_emg_data = all_emg_data[all_emg_data['direction'] == 'forward']
all_qtm_data = all_qtm_data[all_qtm_data['direction'] == 'forward']
```

```python
all_qtm_data_t1 = all_qtm_data[all_qtm_data['target'] == 1]
all_qtm_data_t2 = all_qtm_data[all_qtm_data['target'] == 2]
all_qtm_data_t3 = all_qtm_data[all_qtm_data['target'] == 3]
all_qtm_data_t4 = all_qtm_data[all_qtm_data['target'] == 4]
all_qtm_data_t5 = all_qtm_data[all_qtm_data['target'] == 5]
```

```python
all_emg_data = all_emg_data[all_emg_data['muscle'] != 'Deltoid_Med']
all_emg_data_t1 = all_emg_data[all_emg_data['target'] == 1]
all_emg_data_t2 = all_emg_data[all_emg_data['target'] == 2]
all_emg_data_t3 = all_emg_data[all_emg_data['target'] == 3]
all_emg_data_t4 = all_emg_data[all_emg_data['target'] == 4]
all_emg_data_t5 = all_emg_data[all_emg_data['target'] == 5]
```

```python
## Aggregate trajectories
emg_traces_t1 = figures.aggregate_all_trajectories_qual(all_qtm_data_t1, all_emg_data_t1, unit = 'rad')
emg_traces_t2 = figures.aggregate_all_trajectories_qual(all_qtm_data_t2, all_emg_data_t2, unit = 'rad')
emg_traces_t3 = figures.aggregate_all_trajectories_qual(all_qtm_data_t3, all_emg_data_t3, unit = 'rad')
emg_traces_t4 = figures.aggregate_all_trajectories_qual(all_qtm_data_t4, all_emg_data_t4, unit = 'rad')
emg_traces_t5 = figures.aggregate_all_trajectories_qual(all_qtm_data_t5, all_emg_data_t5, unit = 'rad')
```

```python
# Définition des axes
fig, axes = plt.subplots(3,4, figsize = (4 * pltSizex, 3 * pltSizex), sharex=True, sharey='row', frameon = False, dpi = 300)

# Définition des conditions ("1DDL_O_CB" : 1, not yet tested)
conditions_dict_latex = { 'NE' : 0,
                          'OL' : 1,
                          'CL' : 2,
                          'OLCL' : 3}
conditions_dict = { 'NE' : 0,
                    'OL' : 1,
                    'CL' : 2,
                    'OLCL' : 3}

# Gestion des titres
inv_map = {v: k for k, v in conditions_dict_latex.items()}

for cond in conditions_dict.keys():
    print(cond)
    # Get EMG data
    bic    = emg_traces_t3.get("bic").get(cond)
    bra    = emg_traces_t3.get("bra").get(cond)
    tlat   = emg_traces_t3.get("tlat").get(cond)
    tlng   = emg_traces_t3.get("tlng").get(cond)
    dant   = emg_traces_t3.get("dant").get(cond)
    dpost  = emg_traces_t3.get("dpost").get(cond)
    
    points_mvt = 100
    points_deb_fin = 30
    freq = 3
    order = 5
    s = 13

    ## Upward flexors
    # Get biceps activation for one participant
    bic_av_os     = emg_proc.envelope(abs(np.array(bic.get('mvt')[10:,s])),2000,freq,order)
    bic_sup_os    = emg_proc.envelope(abs(np.array(bic.get('sup')[10:,s])),2000,freq,order)
    bic_inf_os    = emg_proc.envelope(abs(np.array(bic.get('inf')[10:,s])),2000,freq,order)
    # Get brachio-radial activation for one participant
    bra_av_os     = emg_proc.envelope(abs(np.array(bra.get('mvt')[10:,s])),2000,freq,order)
    bra_sup_os    = emg_proc.envelope(abs(np.array(bra.get('sup')[10:,s])),2000,freq,order)
    bra_inf_os    = emg_proc.envelope(abs(np.array(bra.get('inf')[10:,s])),2000,freq,order)
    # Get triceps long head activation for one participant
    tlng_av_os     = emg_proc.envelope(abs(np.array(tlng.get('mvt')[10:,s])),2000,freq,order)
    tlng_sup_os    = emg_proc.envelope(abs(np.array(tlng.get('sup')[10:,s])),2000,freq,order)
    tlng_inf_os    = emg_proc.envelope(abs(np.array(tlng.get('inf')[10:,s])),2000,freq,order)
    # Get triceps lateral head activation for one participant
    tlat_av_os     = emg_proc.envelope(abs(np.array(tlat.get('mvt')[10:,s])),2000,freq,order)
    tlat_sup_os    = emg_proc.envelope(abs(np.array(tlat.get('sup')[10:,s])),2000,freq,order)
    tlat_inf_os    = emg_proc.envelope(abs(np.array(tlat.get('inf')[10:,s])),2000,freq,order)
    # Get deltoid anterior head activation for one participant
    dant_av_os     = emg_proc.envelope(abs(np.array(dant.get('mvt')[10:,s])),2000,freq,order)
    dant_sup_os    = emg_proc.envelope(abs(np.array(dant.get('sup')[10:,s])),2000,freq,order)
    dant_inf_os    = emg_proc.envelope(abs(np.array(dant.get('inf')[10:,s])),2000,freq,order)
    # Get deltoid posterior head activation for one participant
    dpost_av_os     = emg_proc.envelope(abs(np.array(dpost.get('mvt')[10:,s])),2000,freq,order)
    dpost_sup_os    = emg_proc.envelope(abs(np.array(dpost.get('sup')[10:,s])),2000,freq,order)
    dpost_inf_os    = emg_proc.envelope(abs(np.array(dpost.get('inf')[10:,s])),2000,freq,order)
    
    #  Adapt Time vectors
    t_emg     = np.linspace(points_deb_fin, points_mvt + points_deb_fin, len(bic_av_os))
    
    ## Plot phasic flexors and extensors
    # Plot bic up
    axes[0,conditions_dict.get(cond)].plot(t_emg, dant_av_os, linewidth = 5, c = 'blue')
    axes[0,conditions_dict.get(cond)].fill_between(t_emg, dant_inf_os, dant_sup_os, color = 'blue', alpha = 0.15)
    # Plot bra up
    axes[0,conditions_dict.get(cond)].plot(t_emg, dpost_av_os, linewidth = 5, c = 'green')
    axes[0,conditions_dict.get(cond)].fill_between(t_emg, dpost_inf_os, dpost_sup_os, color = 'green', alpha = 0.15)
    # Plot tlng up
    axes[1,conditions_dict.get(cond)].plot(t_emg, bic_av_os, linewidth = 5, c = 'blue')
    axes[1,conditions_dict.get(cond)].fill_between(t_emg, bic_inf_os, bic_sup_os, color = 'blue', alpha = 0.15)
    # Plot tlat up
    axes[1,conditions_dict.get(cond)].plot(t_emg, tlng_av_os, linewidth = 5, c = 'green')
    axes[1,conditions_dict.get(cond)].fill_between(t_emg, tlng_inf_os, tlng_sup_os, color = 'green', alpha = 0.15)
    # Plot tlng up
    axes[2,conditions_dict.get(cond)].plot(t_emg, bra_av_os, linewidth = 5, c = 'blue')
    axes[2,conditions_dict.get(cond)].fill_between(t_emg, bra_inf_os, bra_sup_os, color = 'blue', alpha = 0.15)
    # Plot tlat up
    axes[2,conditions_dict.get(cond)].plot(t_emg, tlat_av_os, linewidth = 5, c = 'green')
    axes[2,conditions_dict.get(cond)].fill_between(t_emg, tlat_inf_os, tlat_sup_os, color = 'green', alpha = 0.15)
    
    
    
axes[0,0].legend(frameon = False, bbox_to_anchor=(1.05, 1.02), fontsize = 19).remove()
axes[0,1].legend(frameon = False, bbox_to_anchor=(1.05, 1.02), fontsize = 19).remove()
    

axes[0,0].set_ylabel(r'Envelope ($\%$ of Max)', fontsize = yTitleFont, labelpad = 10)
axes[0,0].xaxis.set_major_locator(ticker.MaxNLocator(2))
axes[0,0].tick_params(axis='x', which='major', pad=5)
axes[0,0].set_yticks([0, 10, 20], minor=False)
axes[0,0].set_yticklabels([0, 10, 20], size = xLabelFont)
#axes[0,0].set_ylim([-35, 35])

axes[1,0].set_ylabel(r'Envelope ($\%$ of Max)', fontsize = yTitleFont, labelpad = 10)
axes[1,0].xaxis.set_major_locator(ticker.MaxNLocator(2))
axes[1,0].tick_params(axis='x', which='major', pad=5)
axes[1,0].set_yticks([0, 10, 20], minor=False)
axes[1,0].set_yticklabels([0, 10, 20], size = xLabelFont)
#axes[1,0].set_ylim([-35, 35])

axes[2,0].set_ylabel(r'Envelope ($\%$ of Max)', fontsize = yTitleFont, labelpad = 10)
axes[2,0].xaxis.set_major_locator(ticker.MaxNLocator(2))
axes[2,0].tick_params(axis='x', which='major', pad=5)
axes[2,0].set_yticks([0, 4, 8], minor=False)
axes[2,0].set_yticklabels([0, 4, 8], size = xLabelFont)
#axes[2,0].set_ylim([-40, 40])

for axs in axes:
    for ax in axs:
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(4)
        ax.spines['bottom'].set_linewidth(4)
        ax.tick_params(width = 4, length = 12)
        ax.set_xticks([30, 130], minor=False)
        ax.set_xticklabels([0, 100], size = xLabelFont)
        ax.set_xlim([0, 160])

# axes[2].set_xlabel('Time ($\%$ of $\mathrm{\mathbf{MD}}$)', fontsize = yTitleFont, labelpad = 10)
# axes[2].set_ylabel('EMG Activity ($\%$ of Max)', fontsize = yTitleFont, labelpad = 10)
# axes[2].xaxis.set_major_locator(ticker.MaxNLocator(2))
# axes[2].xaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
# axes[2].tick_params(axis='x', which='major', pad=5)
# axes[2].set_xticks([50, 150], minor=False)
# axes[2].set_xticklabels([0, 100], size = xLabelFont)
# axes[2].yaxis.set_major_locator(ticker.MaxNLocator(4))
# axes[2].set_yticklabels(axes[2].get_yticks(), size = xLabelFont)
# axes[2].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
# axes[2].grid(axis='y', zorder = 0)
# axes[2].set_ylim([-20, 30])

fig.text(0.53, -0.02, r'Time ($\%$ of \textbf{MD})', ha='center', size = xTitleFont)

fig.tight_layout()
plt.savefig('./Figures_Unprocessed/smooth_emg_os_lineplots.svg', format='svg')
```
