```python
def set_axis_figAllParams(fig,axes):
    plot_order = ['$T_1$','$T_2$','$T_3$','$T_4$','$T_5$']
    
    axes[0,0].set_title('Flexors', fontsize = yTitleFont)
    axes[0,0].set_xlabel('')
    axes[0,0].set_ylabel('RMS ($\mathrm{\%}$)', fontsize = yTitleFont, labelpad = 10)
    axes[0,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,0].tick_params(axis='x', which='major', pad=5)
    axes[0,0].set_yticks([0,10,20,30])
    axes[0,0].set_yticklabels([0,10,20,30], size = xTitleFont)
    axes[0,0].set_ylim([0, 25])
    
    axes[0,1].set_title('Extensors', fontsize = yTitleFont)
    axes[0,1].set_xlabel('')
    axes[0,1].set_ylabel('', fontsize = yTitleFont, labelpad = 10)
    axes[0,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,1].tick_params(axis='x', which='major', pad=5)
    axes[0,1].set_yticks([0,5,10])
    axes[0,1].set_yticklabels([0,5,10], size = xTitleFont)
    axes[0,1].set_ylim([0, 10])
    
    axes[1,0].set_xlabel('')
    axes[1,0].set_ylabel('RMS ($\mathrm{\%}$)', fontsize = yTitleFont, labelpad = 10)
    axes[1,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,0].tick_params(axis='x', which='major', pad=5)
    axes[1,0].set_yticks([0,5,10,15])
    axes[1,0].set_yticklabels([0,5,10,15], size = xTitleFont)
    axes[1,0].set_ylim([0,18])
    
    axes[1,1].set_xlabel('')
    axes[1,1].set_ylabel('', fontsize = yTitleFont, labelpad = 10)
    axes[1,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,1].tick_params(axis='x', which='major', pad=5)
    axes[1,1].set_yticks([0,5,10,15])
    axes[1,1].set_yticklabels([0,5,10,15], size = xTitleFont)
    axes[1,1].set_ylim([0, 12])
    
    for axs in axes:
        for ax in axs:
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.spines['left'].set_linewidth(4)
            ax.spines['bottom'].set_linewidth(4)
            ax.tick_params(width = 4, length = 12)
            ax.set_xticklabels(plot_order, size = xLabelFont)
            ax.tick_params(axis='x', which='major', pad=5)
            ax.legend().remove()
```

# Relative movement efficiency

```python
## Forward efficiency elbow
# Extensors
emg_eff_elbext_fw = emg_data_elbext_fw[['subject','condition','movement','target','burst\_area','PA','mA']]
emg_eff_elbext_fw = emg_eff_elbext_fw.replace(np.nan,0)
# emg_eff_elbext_fw = emg_eff_elbext_fw.dropna(axis=0)
emg_eff_elbext_fw = emg_eff_elbext_fw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_elbext_fw['ratio_fw_elbext'] = emg_eff_elbext_fw['burst\_area'] / emg_eff_elbext_fw['PA']
# Flexors
emg_eff_elbflex_fw = emg_data_elbflex_fw[['subject','condition','movement','target','deact\_area','PA','mA']]
emg_eff_elbflex_fw = emg_eff_elbflex_fw.replace(np.nan,0)
# emg_eff_elbflex_fw = emg_eff_elbflex_fw.dropna(axis=0)
emg_eff_elbflex_fw = emg_eff_elbflex_fw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_elbflex_fw['ratio_fw_elbflex'] = emg_eff_elbflex_fw['deact\_area'] / emg_eff_elbflex_fw['PA']

## Forward efficiency shoulder
# Extensors
emg_eff_shoext_fw = emg_data_shoext_fw[['subject','condition','movement','target','deact\_area','PA','mA']]
emg_eff_shoext_fw = emg_eff_shoext_fw.replace(np.nan,0)
# emg_eff_shoext_fw = emg_eff_shoext_fw.dropna(axis=0)
emg_eff_shoext_fw = emg_eff_shoext_fw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_shoext_fw['ratio_fw_shoext'] = emg_eff_shoext_fw['deact\_area'] / emg_eff_shoext_fw['PA']
# Flexors
emg_eff_shoflex_fw = emg_data_shoflex_fw[['subject','condition','movement','target','burst\_area','PA','mA']]
emg_eff_shoflex_fw = emg_eff_shoflex_fw.replace(np.nan,0)
# emg_eff_shoflex_fw = emg_eff_shoflex_fw.dropna(axis=0)
emg_eff_shoflex_fw = emg_eff_shoflex_fw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_shoflex_fw['ratio_fw_shoflex'] = emg_eff_shoflex_fw['burst\_area'] / emg_eff_shoflex_fw['PA']

## Merge all dataframes forward
all_eff_merged_fw = emg_eff_elbext_fw.merge(emg_eff_elbflex_fw, on = ['subject','condition','movement','target','PA','mA'])
all_eff_merged_fw = all_eff_merged_fw.merge(emg_eff_shoext_fw, on = ['subject','condition','movement','target','PA','mA'])
all_eff_merged_fw = all_eff_merged_fw.merge(emg_eff_shoflex_fw, on = ['subject','condition','movement','target','PA','mA'])
```

```python
## Backard efficiency elbow
# Extensors
emg_eff_elbext_bw = emg_data_elbext_bw[['subject','condition','movement','target','deact\_area','PA','mA']]
emg_eff_elbext_bw = emg_eff_elbext_bw.replace(np.nan,0)
# emg_eff_elbext_bw = emg_eff_elbext_bw.dropna(axis=0)
emg_eff_elbext_bw = emg_eff_elbext_bw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_elbext_bw['ratio_bw_elbext'] = emg_eff_elbext_bw['deact\_area'] / emg_eff_elbext_bw['PA']
# Flexors
emg_eff_elbflex_bw = emg_data_elbflex_bw[['subject','condition','movement','target','burst\_area','PA','mA']]
emg_eff_elbflex_bw = emg_eff_elbflex_bw.replace(np.nan,0)
#emg_eff_elbflex_bw = emg_eff_elbflex_bw.dropna(axis=0)
emg_eff_elbflex_bw = emg_eff_elbflex_bw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_elbflex_bw['ratio_bw_elbflex'] = emg_eff_elbflex_bw['burst\_area'] / emg_eff_elbflex_bw['PA']

## Backward efficiency shoulder
# Extensors
emg_eff_shoext_bw = emg_data_shoext_bw[['subject','condition','movement','target','burst\_area','PA','mA']]
emg_eff_shoext_bw = emg_eff_shoext_bw.replace(np.nan,0)
# emg_eff_shoext_bw = emg_eff_shoext_bw.dropna(axis=0)
emg_eff_shoext_bw = emg_eff_shoext_bw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_shoext_bw['ratio_bw_shoext'] = emg_eff_shoext_bw['burst\_area'] / emg_eff_shoext_bw['PA']
# Flexors
emg_eff_shoflex_bw = emg_data_shoflex_bw[['subject','condition','movement','target','deact\_area','PA','mA']]
emg_eff_shoflex_bw = emg_eff_shoflex_bw.replace(np.nan,0)
# emg_eff_shoflex_bw = emg_eff_shoflex_bw.dropna(axis=0)
emg_eff_shoflex_bw = emg_eff_shoflex_bw.groupby(['subject','condition','movement','target']).mean().reset_index()
emg_eff_shoflex_bw['ratio_bw_shoflex'] = emg_eff_shoflex_bw['deact\_area'] / emg_eff_shoflex_bw['PA']

## Merge all dataframes
all_eff_merged_bw = emg_eff_elbext_bw.merge(emg_eff_elbflex_bw, on = ['subject','condition','movement','target','PA','mA'])
all_eff_merged_bw = all_eff_merged_bw.merge(emg_eff_shoext_bw, on = ['subject','condition','movement','target','PA','mA'])
all_eff_merged_bw = all_eff_merged_bw.merge(emg_eff_shoflex_bw, on = ['subject','condition','movement','target','PA','mA'])
```

```python
## Compute ratio in the NE condition forward
all_eff_merged_fw_NE = all_eff_merged_fw[all_eff_merged_fw['condition'] == 'NE']
all_eff_merged_fw_NE['ratio_full_NE'] = all_eff_merged_fw_NE['ratio_fw_elbext'] + all_eff_merged_fw_NE['ratio_fw_elbflex'] + all_eff_merged_fw_NE['ratio_fw_shoext'] + all_eff_merged_fw_NE['ratio_fw_shoflex']
all_eff_merged_fw_NE = all_eff_merged_fw_NE[all_eff_merged_fw_NE['ratio_full_NE'] > 0.1]
all_eff_merged_fw_NE = all_eff_merged_fw_NE.groupby(['subject','target']).mean().reset_index()
all_eff_merged_fw_NE = all_eff_merged_fw_NE[['subject','target','ratio_full_NE']]
all_eff_merged_fw_NE = all_eff_merged_fw_NE.dropna(axis = 0)
```

```python
## Compute ratio in the NE condition backward
all_eff_merged_bw_NE = all_eff_merged_bw[all_eff_merged_bw['condition'] == 'NE']
all_eff_merged_bw_NE['ratio_full_NE'] = all_eff_merged_bw_NE['ratio_bw_elbext'] + all_eff_merged_bw_NE['ratio_bw_elbflex'] + all_eff_merged_bw_NE['ratio_bw_shoext'] + all_eff_merged_bw_NE['ratio_bw_shoflex']
all_eff_merged_bw_NE = all_eff_merged_bw_NE[all_eff_merged_bw_NE['ratio_full_NE'] > 0.1]
all_eff_merged_bw_NE = all_eff_merged_bw_NE.groupby(['subject','target']).mean().reset_index()
all_eff_merged_bw_NE = all_eff_merged_bw_NE[['subject','target','ratio_full_NE']]
```

```python
noNE_eff_merged_fw = all_eff_merged_fw[all_eff_merged_fw['condition'] != 'NE']
noNE_eff_merged_fw['ratio_full'] = noNE_eff_merged_fw['ratio_fw_elbext'] + noNE_eff_merged_fw['ratio_fw_elbflex'] + noNE_eff_merged_fw['ratio_fw_shoext'] + noNE_eff_merged_fw['ratio_fw_shoflex']
noNE_eff_merged_fw = noNE_eff_merged_fw[noNE_eff_merged_fw['ratio_full'] > 0.1]

noNE_eff_merged_fw = noNE_eff_merged_fw.groupby(['subject','target','condition']).mean().reset_index()

noNE_eff_merged_fw = noNE_eff_merged_fw.merge(all_eff_merged_fw_NE, on = ['subject','target'])
noNE_eff_merged_fw['rel\_ratio'] = noNE_eff_merged_fw['ratio_full']/noNE_eff_merged_fw['ratio_full_NE'] - 1
noNE_eff_merged_fw = noNE_eff_merged_fw[noNE_eff_merged_fw['rel\_ratio'] > 0.1]
```

```python
noNE_eff_merged_bw = all_eff_merged_bw[all_eff_merged_bw['condition'] != 'NE']
noNE_eff_merged_bw['ratio_full'] = noNE_eff_merged_bw['ratio_bw_elbext'] + noNE_eff_merged_bw['ratio_bw_elbflex'] + noNE_eff_merged_bw['ratio_bw_shoext'] + noNE_eff_merged_bw['ratio_bw_shoflex']
noNE_eff_merged_bw = noNE_eff_merged_bw[noNE_eff_merged_bw['ratio_full'] > 0.1]

noNE_eff_merged_bw = noNE_eff_merged_bw.groupby(['subject','target','condition']).mean().reset_index()

noNE_eff_merged_bw = noNE_eff_merged_bw.merge(all_eff_merged_bw_NE, on = ['subject','target'])
noNE_eff_merged_bw['rel\_ratio'] = noNE_eff_merged_bw['ratio_full']/noNE_eff_merged_bw['ratio_full_NE'] - 1
noNE_eff_merged_bw = noNE_eff_merged_bw[noNE_eff_merged_bw['rel\_ratio'] > 0.1]
```

```python
def set_axis_figRelEff(fig,axes):
    #plot_order = ['$\mathrm{OL}$','$\mathrm{CL}$','$\mathrm{OLCL}$']
    plot_order = ['$T_1$','$T_2$','$T_3$','$T_4$','$T_5$']
    
    axes[0].set_xlabel('', fontsize = yTitleFont, labelpad = 10)
    axes[0].set_title('Forward', fontsize = yTitleFont)
    axes[0].set_ylabel('Efficiency (a.u.)', fontsize = yTitleFont, labelpad = 10)
    axes[0].set_xticklabels(plot_order, size = xLabelFont)
    axes[0].tick_params(axis='x', which='major', pad=5)
    axes[0].set_yticks([0,5,10,15])
    axes[0].set_yticklabels([0,5,10,15], size = xLabelFont)
    axes[0].set_ylim([0, 10.5])
    
    axes[1].set_title('Backward', fontsize = yTitleFont)
    axes[1].set_xlabel('', fontsize = yTitleFont, labelpad = 10)
    axes[1].set_ylabel('', fontsize = yTitleFont, labelpad = 10)
    axes[1].set_xticklabels(plot_order, size = xLabelFont)
    axes[1].tick_params(axis='x', which='major', pad=5)
    
    for ax in axes:
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(4)
        ax.spines['bottom'].set_linewidth(4)
        ax.tick_params(width = 4, length = 12)
        ax.tick_params(axis='x', which='major', pad=5)
        ax.legend(fontsize=xLabelFont*0.75).remove()
```

```python
# hue_order = ['forward','backward']
hue_order = ['OL','CL','OLCL']
cpal = sns.color_palette("dark:#5A9_r", as_cmap=False)
## Figure creation
fig, axes = plt.subplots(1,2,figsize = (2 * pltSizex, 1 * pltSizey), sharex = True, sharey = 'row', dpi = dpi)

sns.barplot(x = 'target', y= 'rel\_ratio', data = noNE_eff_merged_fw, hue = 'condition', hue_order = hue_order, palette=cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = 68, zorder=5, ax = axes[0])

sns.barplot(x = 'target', y= 'rel\_ratio', data = noNE_eff_merged_bw, hue = 'condition', hue_order = hue_order, palette=cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = 68, zorder=5, ax = axes[1])

set_axis_figRelEff(fig,axes)

fig.tight_layout(pad = 3)

plt.savefig('./Figures_Unprocessed/RelativeEfficiency.svg', format='svg')
```
