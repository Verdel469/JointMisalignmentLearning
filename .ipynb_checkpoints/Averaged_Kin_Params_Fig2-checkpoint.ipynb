```python
def set_axis_figAllParams(fig,axes):
    plot_order = ['$T_1$','$T_2$','$T_3$','$T_4$','$T_5$']
    
    axes[0,0].set_xlabel('')
    axes[0,0].set_ylabel('$\mathrm{MD}$ ($\mathrm{s}$)', fontsize = yTitleFont, labelpad = 10)
    axes[0,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,0].tick_params(axis='x', which='major', pad=5)
    axes[0,0].set_yticks([0,1,2])
    axes[0,0].set_yticklabels([0,1,2], size = xTitleFont)
    axes[0,0].set_ylim([0, 2.5])
    
    axes[0,1].set_xlabel('')
    axes[0,1].set_ylabel('Curv. ($\mathrm{cm}$)', fontsize = yTitleFont, labelpad = 10)
    axes[0,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,1].tick_params(axis='x', which='major', pad=5)
    axes[0,1].set_yticks([0.0,0.02,0.04,0.06,0.08])
    axes[0,1].set_yticklabels([0,2,4,6,8], size = xLabelFont)
    axes[0,1].set_ylim([0, 0.082])
    
    #axes[0,2].set_xlabel('')
    #axes[0,2].set_ylabel('$\mathrm{mV}$ ($\mathrm{m.s^{-1}}$)', fontsize = yTitleFont, labelpad = 10)
    #axes[0,2].set_xticklabels(plot_order, size = xTitleFont)
    #axes[0,2].tick_params(axis='x', which='major', pad=5)
    #axes[0,2].set_yticks([0.0,0.2,0.4,0.6])
    #axes[0,2].set_yticklabels([0.0,0.2,0.4,0.6], size = xLabelFont)
    #axes[0,2].set_ylim([0, 0.65])
    
    #axes[0,3].set_xlabel('')
    #axes[0,3].set_ylabel('$\mathrm{mA}$ ($\mathrm{m.s^{-2}}$)', fontsize = yTitleFont, labelpad = 10)
    #axes[0,3].set_xticklabels(plot_order, size = xTitleFont)
    #axes[0,3].tick_params(axis='x', which='major', pad=5)
    #axes[0,3].set_yticks([0,2,4,6])
    #axes[0,3].set_yticklabels([0,2,4,6], size = xLabelFont)
    #axes[0,3].set_ylim([0, 4.5])
    
    axes[1,0].set_xlabel('')
    axes[1,0].set_ylabel('$\mathrm{PV}$ ($\mathrm{m.s^{-1}}$)', fontsize = yTitleFont, labelpad = 10)
    axes[1,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,0].tick_params(axis='x', which='major', pad=5)
    axes[1,0].set_yticks([0.0,0.5,1.0,1.5])
    axes[1,0].set_yticklabels([0.0,0.5,1.0,1.5], size = xLabelFont)
    axes[1,0].set_ylim([0, 1.5])
    
    axes[1,1].set_xlabel('')
    axes[1,1].set_ylabel('$\mathrm{PA}$ ($\mathrm{m.s^{-2}}$)', fontsize = yTitleFont, labelpad = 10)
    axes[1,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,1].tick_params(axis='x', which='major', pad=5)
    axes[1,1].set_yticks([0,2,4,6])
    axes[1,1].set_yticklabels([0,2,4,6], size = xLabelFont)
    axes[1,1].set_ylim([0, 6.5])
    
    #axes[1,2].set_xlabel('')
    #axes[1,2].set_ylabel('$\mathrm{rtPV}$ ($\% \mathrm{MD}$)', fontsize = yTitleFont, labelpad = 10)
    #axes[1,2].set_xticklabels(plot_order, size = xTitleFont)
    #axes[1,2].tick_params(axis='x', which='major', pad=5)
    #axes[1,2].set_yticks([0.0,0.2,0.4])
    #axes[1,2].set_yticklabels([0,20,40], size = xLabelFont)
    #axes[1,2].set_ylim([0, 0.5])
    
    #axes[1,3].set_xlabel('')
    #axes[1,3].set_ylabel('$\mathrm{rtPA}$ ($\% \mathrm{MD}$)', fontsize = yTitleFont, labelpad = 10)
    #axes[1,3].set_xticklabels(plot_order, size = xTitleFont)
    #axes[1,3].tick_params(axis='x', which='major', pad=5)
    #axes[1,3].set_yticks([0.0,0.1,0.2])
    #axes[1,3].set_yticklabels([0,10,20], size = xLabelFont)
    #axes[1,3].set_ylim([0, 0.21])
    
    for axs in axes:
        for ax in axs:
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.spines['left'].set_linewidth(4)
            ax.spines['bottom'].set_linewidth(4)
            ax.tick_params(width = 4, length = 12)
            ax.set_xticklabels(plot_order, size = xLabelFont)
            ax.tick_params(axis='x', which='major', pad=5)
            ax.legend().remove()
```
