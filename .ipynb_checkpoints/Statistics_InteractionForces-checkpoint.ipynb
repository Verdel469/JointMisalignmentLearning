```python
## Load population robot data
all_rob_data = pd.read_csv("./all_params/Transp_2DDL_rob_cut-off-rel-5.csv", encoding = "ISO-8859-1", engine='python')
all_rob_data = all_rob_data[all_rob_data['condition'] != 'NE']
all_rob_data = all_rob_data.groupby(['subject','condition','target','direction']).mean().reset_index()
all_rob_data = all_rob_data[all_rob_data['direction'] == 'forward']
```

# Non-parametric Friedman ANOVAS

```python
# Condition effect Maximum FzA
pg.friedman(all_rob_data, dv = 'mFzA', within = 'condition', subject = 'subject')
```

```python
# Condition effect Mean FzA
pg.friedman(all_rob_data, dv = 'avFzA', within = 'condition', subject = 'subject')
```

```python
# Condition effect Maximum FzFA
pg.friedman(all_rob_data, dv = 'mFzFA', within = 'condition', subject = 'subject')
```

```python
# Condition effect Mean FzFA
pg.friedman(all_rob_data, dv = 'avFzFA', within = 'condition', subject = 'subject')
```
