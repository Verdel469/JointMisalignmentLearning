```python
# Size of one plot and resolution
pltSizex = 10
pltSizey = 10
dpi = 300
# Fontsize for the different parameters
legFont = 55
xTitleFont = 60
yTitleFont = 60
xLabelFont = 55
yLabelFont = 55
# Define linewidth for line plots
lw = 5
# Color for each condition
cmap = colmap.get_cmap('nipy_spectral')
amp15_col = cmap(0.0)
amp30_col = cmap(0.9)
amp45_col  = cmap(0.15)
amp60_col  = cmap(0.5)
# Error bars
ci = 68
```

```python
## Load robot data
all_rob_data = pd.read_csv("./all_params/Transp_2DDL_rob_cut-off-rel-5.csv", encoding = "ISO-8859-1", engine='python')
all_rob_data = all_rob_data[all_rob_data['condition'] != 'NE']
all_rob_data_fw = all_rob_data[all_rob_data['direction'] == 'forward']
all_rob_data_bw = all_rob_data[all_rob_data['direction'] == 'backward']

av_RobData = all_rob_data_fw.groupby(['condition','subject','target']).mean().reset_index()
av_RobData
```

```python
def set_axis_figAllParams(fig,axes):
    plot_order = ['$T_1$','$T_2$','$T_3$','$T_4$','$T_5$']
    
    axes[0,0].set_xlabel('')
    axes[0,0].set_ylabel('Max. $\mid\mathrm{F_{z,A}}\mid$ ($\mathrm{N}$)', fontsize = yTitleFont, labelpad = 10)
    axes[0,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,0].tick_params(axis='x', which='major', pad=5)
    axes[0,0].set_yticks([0,10,20,30])
    axes[0,0].set_yticklabels([0,10,20,30], size = xLabelFont)
    axes[0,0].set_ylim([0, 25])
    
    axes[0,1].set_xlabel('')
    axes[0,1].set_ylabel('Mean $\mid\mathrm{F_{z,A}}\mid$ ($\mathrm{N}$)', fontsize = yTitleFont, labelpad = 10)
    axes[0,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[0,1].set_yticks([0,10,20,30])
    axes[0,1].set_yticklabels(['','','',''], size = xLabelFont)
    axes[0,1].tick_params(axis='x', which='major', pad=5)
    axes[0,1].set_ylim([0, 25])
    
    axes[1,0].set_xlabel('')
    axes[1,0].set_ylabel('Max. $\mid\mathrm{F_{z,FA}}\mid$ ($\mathrm{N}$)', fontsize = yTitleFont, labelpad = 10)
    axes[1,0].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,0].tick_params(axis='x', which='major', pad=5)
    axes[1,0].set_yticks([0,5,10,15,20])
    axes[1,0].set_yticklabels([0,5,10,15,20], size = xLabelFont)
    axes[1,0].set_ylim([0, 21])
    
    axes[1,1].set_xlabel('')
    axes[1,1].set_ylabel('Mean $\mid\mathrm{F_{z,FA}}\mid$ ($\mathrm{N}$)', fontsize = yTitleFont, labelpad = 10)
    axes[1,1].set_xticklabels(plot_order, size = xTitleFont)
    axes[1,1].tick_params(axis='x', which='major', pad=5)
    axes[1,1].set_yticks([0,5,10,15,20])
    axes[1,1].set_yticklabels(['','','','',''], size = xLabelFont)
    axes[1,1].set_ylim([0, 21])
    
    for axs in axes:
        for ax in axs:
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.spines['left'].set_linewidth(4)
            ax.spines['bottom'].set_linewidth(4)
            ax.tick_params(width = 4, length = 12)
            ax.set_xticklabels(plot_order, size = xLabelFont)
            ax.tick_params(axis='x', which='major', pad=5)
            ax.legend().remove()
```

```python
hue_order = ['OL','CL','OLCL']
cpal = sns.color_palette("dark:#5A9_r", as_cmap=False)
cpal = sns.color_palette(['skyblue','dodgerblue','blue'], as_cmap=False)
## Figure creation
fig, axes = plt.subplots(2,2,figsize = (3 * pltSizex, 2 * pltSizey), dpi = dpi)
## Plot MD evolution per block
sns.barplot(x = 'target', y= 'mFzA', data = av_RobData, hue = 'condition', hue_order = hue_order, palette = cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = ci, zorder=5, ax = axes[0,0])
## Plot A evolution per block
sns.barplot(x = 'target', y= 'avFzA', data = av_RobData, hue = 'condition', hue_order = hue_order, palette = cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = ci, zorder=5, ax = axes[0,1])
## Plot PV evolution per block
sns.barplot(x = 'target', y= 'mFzFA', data = av_RobData, hue = 'condition', hue_order = hue_order, palette = cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = ci, zorder=5, ax = axes[1,0])
## Plot PA evolution per block
sns.barplot(x = 'target', y= 'avFzFA', data = av_RobData, hue = 'condition', hue_order = hue_order, palette = cpal, errcolor = 'k', linewidth = 3, edgecolor = 'k', ci = ci, zorder=5, ax = axes[1,1])

set_axis_figAllParams(fig,axes)
fig.tight_layout(pad = 1.0)

plt.savefig('./Figures_Unprocessed/BasicEffortInteractionParams.svg', format='svg')
```
